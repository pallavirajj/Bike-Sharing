### Building model using statsmodel, for the detailed statistics

### Model 1


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Generic function to calculate VIF of variables

def calculateVIF(day):
    vif = pd.DataFrame()
    vif['Features'] = day.columns
    vif['VIF'] = [variance_inflation_factor(day.values, i) for i in range(day.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

```python
# dataframe with RFE selected variables

X_train_rfe = X_train[col]
# calculate VIF

calculateVIF(X_train_rfe)
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Calculating R^2 value for the test dataset
from sklearn.metrics import r2_score
r2= r2_score(y_test, y_pred)
r2
```

```python
# Calucating adjusted R^2 for the test dataset

# n is number of rows in X

n = X_test.shape[0]


# Number of features (predictors, p) is the shape along axis 1
p = X_test.shape[1]

# We find the Adjusted R-squared using the formula

adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```
